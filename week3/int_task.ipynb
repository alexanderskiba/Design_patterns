Существует тоннель длиной 20 км, с одной стороны в него въезжает поезд со скоростью 5 км/ч, с другой стороны влетает муха со скоростью 10 км/ч, муха долетает до поезда и разворачивается, сколько километров пролетит муха

```python
tunnel_length = 20
all_dist = 20
fly_speed = 10
train_speed = 5
fly_dist_acc = 0
train_dist_acc = 0
```

```python
#в данном случае под all_dist будем понимать расстояние, которое каждый раз летит муха до поезда
while tunnel_length - train_dist_acc > 3.552713678800501e-15:
    fly_time = all_dist/(fly_speed + train_speed) # время встречи поезда с мухой
    fly_dist = 2*(fly_time * fly_speed) # расстояние которое пролетит муха до поезда и обратно
    train_dist = 2*(fly_time * train_speed) # Расстояние, которое проедет поезд, пока муха слетает туда обратно
    train_dist_acc += train_dist # поезд все время двигается вперед, необходимо учитывать этот путь для вычисления нового пути мухи
    fly_dist_acc += fly_dist #суммарное расстояние, которое пролетит муха за все время, пока едет поезд
    #теперь пересчитаем расстояние, которое теперь предстоит пролететь мухе
    all_dist = tunnel_length - train_dist_acc
    print(f"Поезд проехал:{train_dist_acc} км, муха пролетела {fly_dist_acc} км")
    
```
